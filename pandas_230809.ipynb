## 결측치 다루기

```python
import pandas as pd
import numpy as np

np.random.seed(0)    # 초전도체 현상, 실험 조건
data = np.random.randint(0, 10, size=(10, 3)).astype(float)
print(data)
```

```python
data[data < 3] = np.nan
print(data)
```

```python
df = pd.DataFrame(data,columns=['col1', 'col2', 'col3'])
df.info()
```

```python
df.isna().sum()
```

```python
# 평균으로 대치 해보기
col3_mean = df['col3'].mean()

df['col3'] = df['col3'].fillna(col3_mean)
df['col2'] = df['col2'].fillna(df['col2'].mean())
df
```

- numpy.where, 조건식 한 번만 적용 <br>
if 조건식, 참일 때, 거짓일 때

```python
# np.where(조건식, 참, np.where(조건식, 참, (np.where(조건식, 참))))
# np.select(다중조건식, 다중결괏값, default ==> else 조건)
```

## shift 기능

```python
temp_df = sales.loc[:, ['datesold', 'price']].copy()
temp_df.head(5)
```

```python
temp_df['shifted_v1'] = temp_df['price'].shift(1, fill_value=0).astype(int)
temp_df['diff'] = temp_df['price'] - temp_df['shifted_v1']
temp_df
```

```python
temp_df['shifted_v2'] = temp_df['price'].shift(2, fill_value=0).astype(int)
temp_df['diff2'] = temp_df['price'] - temp_df['shifted_v2']
temp_df.head(5)
```

```python
temp_df = sales.loc[:, ['datesold', 'price']].copy()

temp_df['shifted_v3'] = temp_df['price'].shift(-1, fill_value=0).astype(int)
temp_df['shifted_v4'] = temp_df['price'].shift(-2, fill_value=0).astype(int)
temp_df.tail()
```

```python
covidtotals[totvars].isnull().sum(axis=0)
totvarsmisscnt = covidtotals[totvars].isnull().sum(axis=1)
covidtotals.loc[totvarsmisscnt>0].T
```

```python
covidtotalsonly = covidtotals.loc[:, totvars]
covidtotalsonly
```

```python
covidtotalsonly = covidtotals.loc[:, totvars]
covidtotalsonly['total_cases_q'] = pd.qcut(covidtotalsonly['total_cases'],
                                          labels=['매우낮음', '낮음', '중간', '높음', '매우높음'],
                                          q = 5, precision=0)

covidtotalsonly['total_deaths_q'] = pd.qcut(covidtotalsonly['total_deaths'],
                                          labels=['매우낮음', '낮음', '중간', '높음', '매우높음'],
                                          q = 5, precision=0)

covidtotalsonly.head(1)
```

```python
pd.crosstab(covidtotalsonly['total_cases_q'], covidtotalsonly['total_deaths_q'])
```
